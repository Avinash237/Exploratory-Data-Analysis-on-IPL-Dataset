**survey of matches played in each venue**

```python
venues_df = data['venue']
venues_df = venues_df.value_counts()
import plotly.graph_objects as go

fig = go.Figure(go.Bar(y=list(venues_df.index),x=list(venues_df.values),orientation='h',marker=dict(color='rgba(255,6,0,0.6)' )))
fig.show()
```
